##Extracting Name, Position, Weight, and Date of Birth

```python
myDataset = pd.DataFrame().assign(nflId=df_players['nflId'], name=df_players['displayName'], position=df_players['Position'], weight=df_players['weight'], age=df_players['birthDate'])
```

```python
myDataset
```

```python
#Convert all heights to feet
import numpy as np

check = df_players['height'].str.split('-', expand=True)
check.columns = ['first', 'second']
check.loc[(check['second'].notnull()), 'first'] = check[check['second'].notnull()]['first'].astype(np.int16) * 12 + check[check['second'].notnull()]['second'].astype(np.int16)

myDataset['height'] = check['first']
myDataset['height'] = myDataset['height'].astype(np.float32)
myDataset['height'] /= 12
```

```python
myDataset = myDataset.dropna()
```

```python
myDataset = myDataset.round({'height': 1})
```

```python
myDataset
```

##Extracting Speed, Distance, Acceleration


```python
df_tracking2018
```

```python
# Speed:

# 2018
max_speed2018 = pd.DataFrame().assign(nflId=df_tracking2018['nflId'], speed_2018=df_tracking2018['s'])
max_speed2018 = max_speed2018.groupby(["nflId"]).max().reset_index()
max_speed2018['nflId'] = max_speed2018['nflId'].astype('int')
max_speed2018 = max_speed2018.round({'speed_2018': 1})

# 2019
max_speed2019 = pd.DataFrame().assign(nflId=df_tracking2019['nflId'], speed_2019=df_tracking2019['s'])
max_speed2019 = max_speed2019.groupby(["nflId"]).max().reset_index()
max_speed2019['nflId'] = max_speed2019['nflId'].astype('int')
max_speed2019 = max_speed2019.round({'speed_2019': 1})

# 2020
max_speed2020 = pd.DataFrame().assign(nflId=df_tracking2020['nflId'], speed_2020=df_tracking2020['s'])
max_speed2020 = max_speed2020.groupby(["nflId"]).max().reset_index()
max_speed2020['nflId'] = max_speed2020['nflId'].astype('int')
max_speed2020 = max_speed2020.round({'speed_2020': 1})
```

```python
max_speed = max_speed2018.merge(max_speed2019.merge(max_speed2020, left_on=['nflId'], right_on=['nflId'], how='right'), left_on=['nflId'], right_on=['nflId'], how='right').dropna()
max_speed
```

```python
# Distance:

#2018
stamina2018 = pd.DataFrame().assign(nflId=df_tracking2018['nflId'], gameId=df_tracking2018['gameId'], dis_2018=df_tracking2018['dis'])
stamina2018 = stamina2018.groupby(["nflId", "gameId"]).sum().reset_index()
stamina2018 = stamina2018.groupby(["nflId"]).mean().reset_index()
stamina2018['nflId'] = stamina2018['nflId'].astype('int')
stamina2018 = stamina2018.round({'dis_2018': 1,})
stamina2018 = stamina2018.drop(["gameId"], axis=1)

#2019
stamina2019 = pd.DataFrame().assign(nflId=df_tracking2019['nflId'], gameId=df_tracking2019['gameId'], dis_2019=df_tracking2019['dis'])
stamina2019 = stamina2019.groupby(["nflId", "gameId"]).sum().reset_index()
stamina2019 = stamina2019.groupby(["nflId"]).mean().reset_index()
stamina2019['nflId'] = stamina2019['nflId'].astype('int')
stamina2019 = stamina2019.round({'dis_2019': 1,})
stamina2019 = stamina2019.drop(["gameId"], axis=1)

#2020
stamina2020 = pd.DataFrame().assign(nflId=df_tracking2020['nflId'], gameId=df_tracking2020['gameId'], dis_2020=df_tracking2020['dis'])
stamina2020 = stamina2020.groupby(["nflId", "gameId"]).sum().reset_index()
stamina2020 = stamina2020.groupby(["nflId"]).mean().reset_index()
stamina2020['nflId'] = stamina2020['nflId'].astype('int')
stamina2020 = stamina2020.round({'dis_2020': 1,})
stamina2020 = stamina2020.drop(["gameId"], axis=1)
```

```python
stamina = stamina2018.merge(stamina2019.merge(stamina2020, left_on=['nflId'], right_on=['nflId'], how='right'), left_on=['nflId'], right_on=['nflId'], how='right').dropna()
stamina
```

```python
# Acceleration:

#2018
Acceleration2018 = pd.DataFrame().assign(nflId=df_tracking2018['nflId'], gameId=df_tracking2018['gameId'], Acc_2018=df_tracking2018['a'])
Acceleration2018 = Acceleration2018.groupby(["nflId", "gameId"]).max().reset_index()
Acceleration2018 = Acceleration2018.groupby(["nflId"]).mean().reset_index()
Acceleration2018['nflId'] = Acceleration2018['nflId'].astype('int')
Acceleration2018 = Acceleration2018.round({'Acc_2018': 1,})
Acceleration2018 = Acceleration2018.drop(["gameId"], axis=1)

#2019
Acceleration2019 = pd.DataFrame().assign(nflId=df_tracking2019['nflId'], gameId=df_tracking2019['gameId'], Acc_2019=df_tracking2019['a'])
Acceleration2019 = Acceleration2019.groupby(["nflId", "gameId"]).max().reset_index()
Acceleration2019 = Acceleration2019.groupby(["nflId"]).mean().reset_index()
Acceleration2019['nflId'] = Acceleration2019['nflId'].astype('int')
Acceleration2019 = Acceleration2019.round({'Acc_2019': 1,})
Acceleration2019 = Acceleration2019.drop(["gameId"], axis=1)

#2020
Acceleration2020 = pd.DataFrame().assign(nflId=df_tracking2020['nflId'], gameId=df_tracking2020['gameId'], Acc_2020=df_tracking2020['a'])
Acceleration2020 = Acceleration2020.groupby(["nflId", "gameId"]).max().reset_index()
Acceleration2020 = Acceleration2020.groupby(["nflId"]).mean().reset_index()
Acceleration2020['nflId'] = Acceleration2020['nflId'].astype('int')
Acceleration2020 = Acceleration2020.round({'Acc_2020': 1,})
Acceleration2020 = Acceleration2020.drop(["gameId"], axis=1)
```

```python
Acceleration = Acceleration2018.merge(Acceleration2019.merge(Acceleration2020, left_on=['nflId'], right_on=['nflId'], how='right'), left_on=['nflId'], right_on=['nflId'], how='right').dropna()
Acceleration
```

##Merging All Extracted Data:

```python
myDataset_final = myDataset.merge( max_speed.merge( stamina.merge(Acceleration, left_on=['nflId'], right_on=['nflId'], how='right'), left_on=['nflId'], right_on=['nflId'], how='right'), left_on=['nflId'], right_on=['nflId'], how='right').dropna()
myDataset_final
```

##Converting Date of Birth to Age

```python
from datetime import datetime, date

today = date.today()

for i in range(0,myDataset_final.shape[0]):
     dob = myDataset_final["age"][i]
     born = datetime.strptime((dob[6:]+"-"+dob[0:2]+"-"+dob[3:5]) if "/" in dob else dob, "%Y-%m-%d").date()
     myDataset_final["age"][i] = today.year - born.year - ((today.month, today.day) < (born.month, born.day))
     
```
